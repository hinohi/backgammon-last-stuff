```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
%matplotlib inline
```

```python
def read(name):
    num = []
    pips = []
    steps = []
    with open(name) as f:
        for line in f:
            a, b = line.split()
            cell = eval(a)
            num.append(sum(cell))
            pips.append(sum(i * c for i, c in enumerate(cell, 1)))
            steps.append(eval(b))
    return np.array(num), np.array(pips), np.array(steps)
```

```python
num, pips, steps = read('results/15.txt')
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax.scatter(pips, steps)
ax.set_xlabel('pips')
ax.set_ylabel('平均必要ステップ')
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax.scatter(pips[num==15], steps[num==15])
ax.set_title('N=15')
ax.set_xlabel('pips')
ax.set_ylabel('平均必要ステップ')
ax.grid()
```
